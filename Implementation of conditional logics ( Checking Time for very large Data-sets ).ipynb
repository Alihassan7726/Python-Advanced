1.if sepal length < 5.1, then label 0;

2.otherwise if sepal width > 3.3 and sepal length < 5.8, then label 1;

3.otherwise if sepal width > 3.3 and petal length > 5.1, then label 2;

4.otherwise if sepal width > 3.3, petal length < 1.6 and either sepal length < 6.4 or petal width < 1.3, then label 3;

5.otherwise if sepal width > 3.3 and either sepal length < 6.4 or petal width < 1.3, then label 4;

6.otherwise if sepal width > 3.3, then label 5;

7.otherwise label 6

#### We will be checking by time 

We'll be using 1.pd.apply 2. .loc[] 3. np.where() 4. np.Select()

```python
%%timeit
def rules(row):
  if row['sepal_length'] < 5.1:
    return 0
  elif row['sepal_width'] > 3.3:
    if row['sepal_length'] < 5.8:
      return 1
    elif row['petal_length'] > 5.1:
      return 2
    elif (row['sepal_length'] < 6.4) or (row['petal_width'] < 1.3):
      if row['petal_length'] < 1.6:
        return 3
      return 4
    return 5
  return 6

iris['label'] = iris.apply(rules, 1)
```
