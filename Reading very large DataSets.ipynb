```python
import pandas as pd
import numpy as np
```

```python
rows = 500000
cols = 150
data = np.random.randint(1,100 , size = (rows , cols))
df = pd.DataFrame(data , columns = ['col_%d'%i for i in range(cols)])
df.head()
```

```python
#pd.options.display.memory_usage(df)
df.info(memory_usage = 'deep')
```

```python
# Creating CSV
file_path = 'big_data.csv'
df.to_csv(file_path , index = False)
```

```python
# Creating hdf5 file
import vaex 
```

```python
v_df = vaex.from_csv(file_path , convert = True , chunk_size = 5_00_00)
v_df
```

```python
type(v_df)
```

```python
v_df = vaex.open('big_data.csv.hdf5')
v_df
```

```python
type(v_df)
```

```python
v_df['col_arb'] = v_df.col_0 * v_df.col_1
v_df['col_arb']
```

# performing some calculation

```python
v_df[v_df.col_0 > 50]
```

```python
print(v_df['col_0'].mean(progress = 'widget'))
#print(v_df['col_0'].median(progress = 'widget')) # Can't use in this way
#print(v_df['col_0'].mode(progress = 'widget'))   # Can'nt use inthis way
print(v_df['col_0'].max(progress = 'widget'))
print(v_df['col_0'].min(progress = 'widget'))
```

```python
vaex.agg.mean(v_df.col_0)
```

# Groupby operation


```python
%%time
v_df.groupby(by = v_df.col_1 , agg = vaex.agg.mean(v_df.col_0))
```

```python
%%time
v_df.groupby(by = v_df.col_1 , agg = 'mean') # We can also give 'count' instead of mean
```
